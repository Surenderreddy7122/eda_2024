**BAR CHART USING SEABORN**
- we can plot bar chart using with many packages 
- one package is matplotlib
- another package is seaborn
- in matplotlib the method is **plt.bar**
- in seaborn the method is **countplot**
- in matplotlib 
   - we take the raw data
   - we will create our frequency table i.e dataframe
   - we will pass the column1,column2 and dataframe name 
- in seaborn count plot
   - it take direct raw data 
   - we will provide column name
   - we will provide main dataframe name

```python
# continent column ==== continent_df====plt.bar
# continent column directly === visadf
import seaborn as sns
plt.figure(figsize=(8,5))
plt.title("bar chart")
plt.xlabel('continents')
plt.ylabel('number of applicants')

order_continent=['Asia','Europe','North America','South America','Africa','Ocenia']

sns.countplot(data=visa_df,x='continent',order=order_continent,color='red')
plt.savefig('continents_bar_chart_using_seaborn.jpg')
plt.show()
```

**PIE CHART**

- PIE chart is a 360 degree view
- it is matplotlib package 
- it will take key and values 
- that will get fro value counts

```python
c_df_key=visa_df['continent'].value_counts().keys()
c_df_key
```

```python
c_df_values=visa_df['continent'].value_counts().values
c_df_values
```

```python
plt.pie(c_df_values,explode=[0.10,0.10,0.10,0.10,0.10,0.10],
        labels=c_df_key,autopct="%0.2f%%",radius=2,
       startangle=90)
plt.show()
```
